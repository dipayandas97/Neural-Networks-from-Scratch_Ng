```
data_frame = pd.read_csv(io.BytesIO(dataset['breast_cancer_MB.csv']))
data = data_frame.values
```

```
X = data[:,2:-1].reshape(30, 569).astype('float64')
Y = data[:,1].reshape(1, 569)

for row in range(X.shape[0]):
  X[row,:] = (X[row,:] - np.mean(X[row,:])) / np.std(X[row,:])

Y[ Y == 'M'] = int(0.0)
Y[ Y == 'B'] = int(1.0)
Y = Y.astype('float64')

n = X.shape[0] # number of features 30
m = X.shape[1] # number of training examples 569

np.max(X[1,:])
```

```
def sigmoid(a):
  return 1 /( 1 + np.exp(-a))
```

```
def del_sigmoid(a):
  return sigmoid(a)*(1 - sigmoid(a))
```

```
def feed_forward(W,X,B):
  Z = np.dot(W, X) + B
  return Z  
```

```
N = [30, 50, 1] # number of neurons in each layer from 0 to 2

W1 = np.random.rand(N[1], N[0])
B1 = np.random.rand(N[1]).reshape(N[1],1)

W2 = np.random.rand(N[2], N[1])
B2 = np.random.rand(N[2]).reshape(N[2],1)

lr = 0.05
```

```
epochs = 10000

for epoch in range(epochs):
  # Feed forward

  Z1 = feed_forward(W1,X,B1)
  A1 = sigmoid(Z1)
  Z2 = feed_forward(W2,A1,B2)
  A2 = sigmoid(Z2)

  #back prop

  dZ_2 = A2 - Y
  dW_2 = np.dot(dZ_2, A1.T) / m  
  dB_2 = np.sum(dZ_2, axis = 1, keepdims = True) / m

  dZ_1 = np.dot(W2.T,dZ_2) * del_sigmoid(Z2)
  dW_1 = np.dot(dZ_1, X.T) / m
  dB_1 = np.sum(dZ_1, axis = 1, keepdims = True) / m
  
  W1 = W1 - ( lr * dW_1 )
  W2 = W2 - ( lr * dW_2 )
  B1 = B1 - ( lr * dB_1 )
  B2 = B2 - ( lr * dB_2 )
  
  if epoch%(0.1*epochs) == 0: print('epochs done: ',epoch)
```

```
def evaluate(inp):
  z1 = feed_forward(W1,inp,B1)
  a1 = sigmoid(z1)

  z2 = feed_forward(W2,a1,B2)
  a2 = sigmoid(z2)
  return a2
```

```
score = 0.0

for i in range(m):
  inp =  X[:,i].reshape(30,1) 

  if evaluate(inp) < 0.5 and Y[0,i]==0: 
    score+=1
  elif evaluate(inp)>0.5 and Y[0,i]==1:
    score+=1

score /= m
score*100
    
```

```
model = Sequential()
model.add(Dense(50, activation = 'sigmoid', input_dim = 30))
model.add(Dense(1, activation = 'sigmoid'))
model.compile(loss = 'binary_crossentropy', 
              optimizer = 'adam', metrics = ['accuracy'])
```

```
x_new = X.T
y_new = Y.T
```

```
model.fit(x_new,y_new, epochs = 1000, verbose = 1)
```
